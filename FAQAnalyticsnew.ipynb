```python
stop=stopwords.words("english")
```

```python
df["question_trans"]=df["question_trans"].apply(lambda x:[lem.lemmatize(item) for item in x if item not in stop])
```

```python
df["question_trans"]=df["question_trans"].apply(lambda x :" ".join(x))
```

```python
df["question_trans_noun"]=df["question_trans"].apply(lambda x :[item for item in nltk.word_tokenize(x) ] )
```

```python
df["question_trans_noun"]=df["question_trans_noun"].apply(lambda x:[nltk.pos_tag(x) for item in x])
```
